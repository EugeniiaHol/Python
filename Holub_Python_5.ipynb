```python
# Завантажуємо текстовий файл з даними

with open('orders.txt', encoding='utf8') as f:
    data = f.read().split('\n\n')
#print(data)

# Перетворюємо дані з текстового файлу на списки чеків зі списками покупок
# відповідно до роздільників

orders = [order.split('@@@') for order in data]
#print(orders)

# виводимо на друк кількість чеків, інформація про які, містилась у файлі з даними
print ("Total orders:  ", len(orders))

# перевіряємо кількість покупок за перршим чеком
print (len(orders[0]))
```

```python
# 2. Код для розрахунку підтримки та впевненості для кожної пари продуктів
```

```python
from collections import defaultdict
```

```python
# 2.а. створюємо 2 словника для підрахунку кожного окремого продукуту 
#      та підрахунку пар продуктів
```

```python
# Визначаємо частоту появи продуктів в чеках та визначаємо скільки усього
# унікальних продуктів

# Створюємо словник з нульовими ключами, значеннями, що зустрічаються вперше
product_count = defaultdict(lambda:0)
product_pair_cont = defaultdict(int)

for order in orders:
    for p in order:
        product_count[p]+=1
# Перебір пар у чеку
    for i, p1 in  enumerate(order[:-1]):
        for p2 in order[i + 1:]:
            product_pair_cont[(p1,p2)]+=1
            product_pair_cont[(p2,p1)]+=1
# Друкуємо кількість різних товарів за сумою чеків
print(len(product_count))
```

```python
# Виводимо на друк частоту появи кожного з товарів

dict(product_count)
```

```python
# 2.с. Виводимо кількість різноманітних пар товарів по чеку 
#      та сумарну кількість однакових пар

dict(product_pair_cont)
```

```python
# 2.d. Розраховуємо довіру та генеруємо новий словник з парами та відповідними значеннями довіри

confidence = {
    (p1,p2): count / product_count[p1] for (p1,p2), count in product_pair_cont.items()
}

confidence
```

```python
# 2.g. Формуємо функцію,щодає встановити мінімальні межі довіри 
#      та підтримки та вивести відфільтровані дані у вказаному форматі.
```

```python
def print_conclusion(min_confidence,min_support,orders_list):
    len_orders_list = len(orders_list)
    for (p1,p2), count in product_pair_cont.items():
        support = count / len_orders_list
        conf = confidence[(p1,p2)]
        if conf >= min_confidence and support >= min_support:
            print (f'{p1} => {p2} ({conf * 100:.2f}% confidence), {support * 100:.2f}% support')
```

```python
print_conclusion(0.05,0.01,orders)
```
