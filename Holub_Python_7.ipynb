2. 1. Визначення рекордів для чоловіків по різним категоріям в різних видах змагань 

```python
df_m = df[df['sex']=='M']
df_m = df_m.groupby('division',dropna=False).agg(max_best3squat_kg=('best3squat_kg', 'max'),
                                  max_best3bench_kg=('best3bench_kg','max'), 
                                  max_best3deadlift_kg=('best3deadlift_kg','max'))
print('Рекорди для чоловіків по 3-м видам змагань:  ')
print(df_m)
```

2. 2. Визначення рекордів для жінок по різним категоріям в різних видах змагань 

```python
df_f = df[df['sex']=='F']
df_f = df_f.groupby('division',dropna=False).agg(max_best3squat_kg=('best3squat_kg', 'max'),
                                  max_best3bench_kg=('best3bench_kg','max'), 
                                  max_best3deadlift_kg=('best3deadlift_kg','max'))
print('Рекорди для жінок по 3-м видам змагань:  ')
print(df_f)
```

3. 1. Визначення загальної кількості перших місцьдля кожного з учасників

```python
df_v = df.dropna()
df_v = df_v.loc[:,['name','place']]
df_v = df_v[df_v['place'] == '1']
df_v = df_v.groupby('name').count()
df_v = df_v.sort_values(by = ['place'],ascending=False)
df_v = df_v.rename(columns={'place':'total_wins','name': 'unique'})
print('Кількість перших місць, що займав спортсмен суммарно за усіма категоріями: ')
df_v
```

3. 2. Визначення суми перших місць для учасників в певних категоріях

```python
df_for_agg = df.dropna()
df_for_agg = df_for_agg[['name','division','sex','place']][df_for_agg.place=='1']
df_for_agg_1 = df_for_agg.groupby(['division','sex','name']).agg(total_wins=('place', 'count'))
print('Кількість перших місць, що займали спортсмени суммарно за визначеною категорією: ')
df_for_agg_1
```

4. Визначення учасників з максимальною кількістю перших місць для кожної категорії

```python
import itertools

loaded_df = pd.read_csv('ipf_lifts.csv')

df_for_agg = loaded_df.dropna()
df_for_agg = df_for_agg[['name','division','sex','place']][df_for_agg.place=='1']
df_for_agg_1 = df_for_agg.groupby(['division','sex','name']).agg(total_wins=('place', 'count'))
df_for_agg_1

result = pd.DataFrame()

for division,sex in itertools.product(df_for_agg.division.unique(),df_for_agg.sex.unique()):
    
    first_place_name = df_for_agg_1.loc[(str(division),str(sex)),:].idxmax()[0]
    
    num = df_for_agg_1.loc[(str(division),str(sex),str(first_place_name)),:]
    
    result_1=pd.DataFrame({'division': division,
                           'sex': sex,
                           'name':first_place_name,
                           'number_win':num})
    
    result = pd.concat([result,result_1])
result.reset_index(drop=True, inplace=True)   
result
```
