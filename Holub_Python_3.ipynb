```python
# Визначення змінних
num = int(input("Введіть кількітьсть студентів відмінну від 0:  "))
my_list = []
my_list_bal = []
my_list_rez = []
my_list_z = []
my_list_n = []
zarah_bal = []
nezar_bal = []

# Введення даних
if num==0:
    print("Кількітьсть студентів 0 - це неприпустимо для розв'язання задачі")
else:
    for i in range(1,num+1):
        nome = str(input("Введіть прізвище студента %s:  "%i)) 
        if nome=="":
            print(f"Введіть непусте значення прізвища студента")
            
        else:
            my_list.append(nome)
    if len(my_list)==num:
        print (my_list) 
    else:
        print ("Кількість введерних стундентів не відповідає замовленому розміру списка,  отже вводьте бали тільки для введених студентів") 
        
    for j in range(1,len(my_list)+1):
        bal = int(input("Введіть бал студента %s:  "%j))
        my_list_bal.append(bal)
    if len(my_list_bal)==num:
        print (my_list_bal)
    else:
        print ("Кількість введених балів стундентів не відповідає замовленому розміру списка") 
        
    for k in range(1,len(my_list)+1):
        res = input("Введіть висновок викладача (Здав - 'З' чи Не задав - 'Н') для студента %s:  "%k)
        my_list_rez.append(res)
    if len(my_list_rez)==num:
          print (my_list_rez)
    else:
        print ("Кількість введених висновків не відповідає замовленому розміру списка")
        
# Упорядкування даних
ml = tuple(my_list)
mlb = tuple(my_list_bal)
mlr = tuple(my_list_rez)
rez_ekz=zip(ml,mlb,mlr)
rez_ekz_sort = sorted(rez_ekz, key = lambda a: a[2])

print("Stud","Bal","Stat", sep="\t")
for element in rez_ekz_sort:
    print(element[0], element[1],element[2], sep="\t")

# Групування даних
for element in rez_ekz_sort:
    if element[2]=="З":
        my_list_z.append(element)
    else:
        my_list_n.append(element)

# Виведення згрупованого списку студентів, яким Зараховано
print("Зараховано, список студентів:")                        
print("Stud","Bal","Stat", sep="\t")
for element in  my_list_z:
     print(element[0], element[1],element[2], sep="\t")
        
# Формування списку балів,що відповідають стану "Зараховано"
for element in  my_list_z:
    zarah_bal.append(element[1])

# Виведення згрупованого списку студентів, яким Не зараховано
print("Не зараховано, список студентів:")    
print("Stud","Bal","Stat", sep="\t")
for element in  my_list_n:
    print(element[0], element[1],element[2], sep="\t")

# Формування списку балів,що відповідають стану "Не зараховано"
for element in  my_list_n:
    nezar_bal.append(element[1])

#  Логіка виведення висновку  
if len(nezar_bal) > 0: 
    n = max(list(nezar_bal)) # Агрегація (максимум)
    if len(zarah_bal) > 0: 
        m = min(list(zarah_bal)) # Агрегація (мінімум)
        if n < m:
            print("Викладач виставив залік коректно")
            print(f"Границя міститься між {n} і {m}")
        else:
            print("Викладач виставив залік некоректно")  
    else: 
        print("Студентів, яким зарахували екзамен, немає")        
else:
    print("Студентів, яким не зарахували екзамен, немає")

```
