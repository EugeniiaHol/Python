```python
# 1. В качестве входящих данных запросите целое число. Если оно делится на 3,
# выведите «foo»; если оно делится на 5, выведите «bar»; если оно делится на оба,
# выведите «ham» (а не «foo» или «bar»).

x = int(input("Ведіть ціле число: "))
if x%3==0 and x%5==0:
    print("ham")
elif x%3==0:
    print("foo")
elif x%5==0:
    print("bar")
else:
    print(x)
```

```python
# 2. В качестве входящих данных запросите два числа и выведите какое из них
# меньше и какое большее.

x1 = float(input("Ведіть перше число: "))
x2 = float(input("Ведіть друге число: "))
if x1>x2:
    print("Перше число більше, а друге - меньше")
elif x1<x2:
    print("Перше число меньше, а друге - більше")
else:
    print("Числа рівні")
```

```python
# 5. Сыграйте в игру Fizz-Buzz: выведите все числа от 1 до 100; если число делится на # 3, 
# вместо числа выведите «fizz». Если оно делится на 5, вместо числа выведите
# “buzz”. Если оно делится на оба, выведите «fizz buzz» вместо числа.

for i in range(1,101):
    if (i%5)==0 and (i%3)==0:
        print("fizz buzz")
    elif (i%3)==0:
        print("fizz")
    elif (i%5)==0:
        print("buzz")
    else:
        print(i)
```
